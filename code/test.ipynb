```python
low_bound = (100, 200)
mid_bound = (200, 300)
high_bound = (300, 400)

def check_canny(img_path):
    fig = plt.figure(figsize=(10, 7))

    original = cv2.resize(cv2.imread(img_path), (200, 200), interpolation = cv2.INTER_AREA)
    
#     kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
#     original = cv2.filter2D(original, -1, kernel)
    
    wide = cv2.Canny(original, low_bound[0], low_bound[1])
    
    mid = cv2.Canny(original, mid_bound[0], mid_bound[1])
    
    tight = cv2.Canny(original, high_bound[0], high_bound[1])

    fig.add_subplot(1, 4, 1) # original image
    plt.imshow(original)
    plt.axis('off')
#     plt.title('Original')

    fig.add_subplot(1, 4, 2) # wide
    plt.imshow(wide)
    plt.axis('off')
#     plt.title('Small-Scale')

    fig.add_subplot(1, 4, 3) # mid
    plt.imshow(mid)
    plt.axis('off')
#     plt.title('Mid-Scale')

    fig.add_subplot(1, 4, 4) # tight
    plt.imshow(tight)
    plt.axis('off')
#     plt.title('Large-Scale')
    
```

```python
check_canny('data/categorized_images/Plant/plant_15.jpg')
```
